### 17. Gráfico de Setores de Grupos Sanguíneos
##### Construa um gráfico de setores que mostre a distribuição dos grupos sanguíneos do Exercício 29, Seção 2-2.

```python
# importando bibliotecas
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Versão bibliotecas
print(f"Versão Pandas: {pd.__version__}")
print(f"Versão Matplotlib: {matplotlib.__version__}")
print(f"Versão Seaborn: {sns.__version__}")
```

```python
# valores do exercício
valores = ['A', 'A', 'A', 'A', 'A', 'A', 'A', 'A', 'A', 'A', 'A', 'A', 'A', 'A', 'A', 'A', 'A', 'A', 'A', 'A',
          'O', 'O', 'O', 'O', 'O', 'O', 'O', 'O', 'O', 'O', 'O', 'O', 'O', 'O', 'O', 'O', 'O', 'O', 'O', 'O', 'O', 'O',
          'B', 'B', 'B', 'B',
          'AB', 'AB', 'AB']
```

```python
# gerando frequência dos valores
freq = pd.Series(valores).value_counts()
```

```python
freq
```

```python
# categorizar frequências em classes
classe = pd.Series({
    "Tipo A": freq['A'],
    "Tipo B": freq['B'],
    "Tipo AB": freq['AB'],
    "Tipo O": freq['O']
}).sort_values(ascending = False)
```

```python
# gerando labels para o gráfico
labels = []
for label in classe.index:
    labels.append(label)
```

```python
# plotando gráfico de setores
fig, ax = plt.subplots()

#utilizando paleta de cores do seaborn
cores = sns.color_palette('RdBu')

ax.pie(x = freq, labels = labels,
      startangle = 90, autopct= '%1.1f%%',
      wedgeprops={'edgecolor': 'black', 'linewidth': 0.2, 'linestyle': '-'},
      textprops={'fontsize': 10}, colors = cores, labeldistance = 1.2, shadow = True)

plt.show()
```
