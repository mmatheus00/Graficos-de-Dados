### construa um diagrama de ramo-e-folhas das quantidades de estrôncio-90.

```python
# importando bibliotecas
import pandas as pd
```

```python
# valores do exercício
valores = [155, 142, 149, 130, 151, 163, 151, 142, 156, 133, 138, 161, 128, 144, 172, 137, 151, 166, 147, 163,
           145, 116, 136, 158, 114, 165, 169, 145, 150, 150, 150, 158, 151, 145, 152, 140, 170, 129, 188, 156]
```

```python
# função para gerar o diagrama de ramo-e-folhas
def text_stem_leaf_plot(data):
    """
    Generates a text-based stem-and-leaf plot from a list or Pandas Series of numerical data.
    """
    # Ensure data is sorted and converted to a list
    data = sorted([int(x) for x in data if pd.notna(x)])

    if not data:
        print("No valid data to plot.")
        return

    # Determine stem unit (e.g., tens place)
    # For simplicity, assumes data is integers and stems are tens digits
    min_val = data[0]
    max_val = data[-1]

    stems = {}
    for x in data:
        stem = x // 10  # Integer division to get the tens digit
        leaf = x % 10   # Modulo to get the units digit
        if stem not in stems:
            stems[stem] = []
        stems[stem].append(leaf)

    # Print the stem-and-leaf plot
    print("Stem-and-Leaf Plot:")
    print("-" * 20)

    # Sort stems for ordered output
    sorted_stems = sorted(stems.keys())

    for stem_val in sorted_stems:
        leaves = sorted(stems[stem_val])
        print(f"{stem_val} | {''.join(map(str, leaves))}")

    print("-" * 20)
```

```python
text_stem_leaf_plot(valores)
```

## O que o diagrama sugere sobre essas quantidades?

1. O formato da distribuição dos dados se assemelha a uma distribuição normal.
2. O centro dos dados está entre os valores 140 e 158.
