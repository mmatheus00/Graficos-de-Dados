### 20. Gráfico de Setores de Descarrilamentos de Trens
#### Construa um gráfico de setores que mostre a distribuição dos descarrilamentos de trens do Exercício 30, Seção 2-2.

```python
# importando bibliotecas
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Versão bibliotecas
print(f"Versão Pandas: {pd.__version__}")
print(f"Versão Matplotlib: {matplotlib.__version__}")
print(f"Versão Seaborn: {sns.__version__}")
```

```python
# valores do exercício
valores = ['E', 'E', 'E', 'E', 'E', 'E', 'E', 'E', 'E',
          'H', 'H', 'H', 'H', 'H', 'H', 'H', 'H', 'H', 'H', 'H', 'H',
          'O', 'O', 'O', 'O', 'O', 'O',
          'T', 'T', 'T', 'T', 'T', 'T', 'T', 'T', 'T', 'T', 'T', 'T', 'T', 'T', 'T', 'T', 'T', 'T', 'T', 'T', 'T', 'T', 'T',]
```

```python
# gerar frequência de valores
freq = pd.Series(valores).value_counts()
freq
```

```python
# categorizar frequências em classes
classe = pd.Series({
    "(E) Equipamento Defeituoso": freq['E'],
    "(H) Falha Humana": freq['H'],
    "(O) Outras Causas": freq['O'],
    "(T) Trilho Ruim": freq['T']
}).sort_values(ascending = False)
```

```python
# gerando labels para o gráfico
labels = []
for label in classe.index:
    labels.append(label)
```

```python
# plotando gráfico de setores
fig, ax = plt.subplots()

#utilizando paleta de cores do seaborn
cores = sns.color_palette('PuOr_r')

ax.pie(x = freq, labels = labels,
      startangle = 90, autopct= '%1.1f%%',
      wedgeprops={'edgecolor': 'black', 'linewidth': 0.2, 'linestyle': '-'},
      textprops={'fontsize': 10}, colors = cores, labeldistance = 1.2, shadow = True)

plt.show()
```
