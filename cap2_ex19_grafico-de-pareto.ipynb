### 19. Gráfico de Pareto de Descarrilamentos de Trens
#### Construa um gráfico de Pareto que mostre a distribuição dos descarrilamentos de trens do Exercício 30, Seção 2-2.

```python
# importando bibliotecas
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Versão bibliotecas
print(f"Versão Pandas: {pd.__version__}")
print(f"Versão Matplotlib: {matplotlib.__version__}")
print(f"Versão Seaborn: {sns.__version__}")
```

```python
# valores do exercício
valores = ['E', 'E', 'E', 'E', 'E', 'E', 'E', 'E', 'E',
          'H', 'H', 'H', 'H', 'H', 'H', 'H', 'H', 'H', 'H', 'H', 'H',
          'O', 'O', 'O', 'O', 'O', 'O',
          'T', 'T', 'T', 'T', 'T', 'T', 'T', 'T', 'T', 'T', 'T', 'T', 'T', 'T', 'T', 'T', 'T', 'T', 'T', 'T', 'T', 'T', 'T',]
```

```python
# gerar frequência de valores
freq = pd.Series(valores).value_counts()
freq
```

```python
# categorizar frequências em classes
classe = pd.Series({
    "(E) Equipamento Defeituoso": freq['E'],
    "(H) Falha Humana": freq['H'],
    "(O) Outras Causas": freq['O'],
    "(T) Trilho Ruim": freq['T']
}).sort_values(ascending = False)
```

```python
# gerando labels para o gráfico
labels = []
for label in classe.index:
    labels.append(label)
```

```python
# plotando gráfico de pareto
sns.catplot(x = labels, y = freq, kind="bar",
           edgecolor = "black", width = 0.5, palette='PuOr_r', hue = labels)

plt.ylabel("Frequência")
plt.xlabel("Causas para Descarrilhamento de Trens")
plt.xticks(labels, ha = 'center', rotation = 45)
plt.ylim(0, 25)
plt.show()
```
