```python
# valores exercício
valores = [0.27, 1.41, 2.19, 2.83, 2.19, 1.81, 0.85, 3.05, 3.42, 2.10, 2.93, 2.44, 2.17,
           1.41, 2.00, 0.93, 2.97, 2.04, 0.65, 2.13, 0.63, 1.53, 4.69, 0.15, 1.45, 2.68,
           3.53, 1.49, 2.31, 0.92, 0.89, 0.80, 0.72, 2.66, 4.37, 0.92, 1.40, 1.45, 1.68,
           1.53, 1.44, 1.44, 1.36, 0.38, 1.74, 2.35, 2.30, 1.14, 2.88, 2.13, 5.28, 1.48,
           3.36, 2.83, 2.87, 2.96, 1.61, 1.58, 1.15, 1.28, 0.58, 0.74]
```

```python
# função modificada para números float
def text_stem_leaf_plot(data):
    """
    Generates a text-based stem-and-leaf plot from a list of numerical data.
    Adapted for decimal numbers with one decimal place.
    """
    # Ensure data is sorted and handle missing values
    data = sorted([float(x) for x in data if pd.notna(x)])

    if not data:
        print("No valid data to plot.")
        return

    # Multiply by 10 to work with integers while preserving decimal information
    scaled_data = [int(x * 10) for x in data]

    stems = {}
    for x in scaled_data:
        stem = x // 10  # Integer division to get the stem (integer part)
        leaf = x % 10   # Modulo to get the leaf (decimal part)
        if stem not in stems:
            stems[stem] = []
        stems[stem].append(leaf)

    # Print the stem-and-leaf plot
    print("Stem-and-Leaf Plot:")
    print("-" * 25)
    print("Stem | Leaves")
    print("-" * 25)

    # Sort stems for ordered output
    sorted_stems = sorted(stems.keys())

    for stem_val in sorted_stems:
        leaves = sorted(stems[stem_val])
        # Format leaves with spaces between them for better readability
        leaves_str = ' '.join(map(str, leaves))
        print(f"{stem_val:4} | {leaves_str}")

    print("-" * 25)
    print("Note: Stem represents integer part, Leaf represents decimal part")
    print("Example: 1 | 4 represents 1.4")
```

```python
text_stem_leaf_plot(valores)
```

### O que o diagrama de ramo-e-folhas sugere sobre a distribuição dos pesos?

Observando o diagrama, vemos que a distribuição dos dados é assimétrica à direita (ou assimetria positiva). Ou seja, o descarte semanal de resíduos plásticos por família geralmente é de até 2.9 lbs (~1,3 kg), sendo que valores superiores a esse são significativamente menos frequentes.
