### 18. Gráfico de Pareto de Grupos Sanguíneos
#### Construa um gráfico de Pareto que mostre a distribuição dos grupos sanguíneos do Exercício 29, Seção 2-2.

```python
# importando bibliotecas
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Versão bibliotecas
print(f"Versão Pandas: {pd.__version__}")
print(f"Versão Matplotlib: {matplotlib.__version__}")
print(f"Versão Seaborn: {sns.__version__}")
```

```python
# valores do exercício
valores = ['A', 'A', 'A', 'A', 'A', 'A', 'A', 'A', 'A', 'A', 'A', 'A', 'A', 'A', 'A', 'A', 'A', 'A', 'A', 'A',
          'O', 'O', 'O', 'O', 'O', 'O', 'O', 'O', 'O', 'O', 'O', 'O', 'O', 'O', 'O', 'O', 'O', 'O', 'O', 'O', 'O', 'O',
          'B', 'B', 'B', 'B',
          'AB', 'AB', 'AB']
```

```python
# gerando frequência dos valores
freq = pd.Series(valores).value_counts()
freq
```

```python
# categorizar frequências em classes
classe = pd.Series({
    "Tipo A": freq['A'],
    "Tipo B": freq['B'],
    "Tipo AB": freq['AB'],
    "Tipo O": freq['O']
}).sort_values(ascending = False)
```

```python
classe
```

```python
# gerando labels para o gráfico
labels = []
for label in classe.index:
    labels.append(label)
```

```python
# deixando os valores ordenados para construção do gráfico
pair = sorted(zip(freq, labels), reverse = True)
freq, labels = zip(*pair)
```

```python
# plotando gráfico de pareto
sns.catplot(x = labels, y = freq, kind="bar",
           edgecolor = "black", width = 0.5, palette='RdBu', hue = labels)

plt.ylabel("Frequência")
plt.xlabel("Grupos Sanguíneos")
plt.xticks(labels, ha = 'center')
plt.ylim(0, 25)
plt.show()
```
