### 13. Gráfico de Pareto para Matrículas na Graduação
#### A Tabela 2-9 (com base em dados do U.S. National Center for Education Statistics) mostra a distribuição de matrículas de estudantes na graduação de faculdades. Construa um gráfico de Pareto para os dados na Tabela 2-9.

```python
# importando bibliotecas
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Versão bibliotecas
print(f"Versão Pandas: {pd.__version__}")
print(f"Versão Matplotlib: {matplotlib.__version__}")
print(f"Versão Seaborn: {sns.__version__}")
```

```python
# valores tabela 2-9
faculdade = ["Pública de 2 anos", "Pública de 4 anos", "Particular de 2 anos", "Particular de 4 anos"]
freq = [36.8, 40.0, 1.6, 21.9]
```

```python
# deixando os valores ordenados para construção do gráfico
pair = sorted(zip(freq, faculdade), reverse = True)
freq, faculdade = zip(*pair)
```

```python
# plotando gráfico de pareto
sns.catplot(x = faculdade, y = freq, kind="bar",
           edgecolor = "black", width = 0.5, palette='inferno', hue = faculdade)

sns.color_palette(palette='inferno')
plt.xticks(faculdade, rotation = 45)
plt.show()
```
