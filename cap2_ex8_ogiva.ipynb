## Exercício 8. Ogiva
### Construa um diagrama de pontos das quantidades de estrôncio-90. Para o eixo horizontal, use as fronteiras de classe correspondentes aos limites de classe dados no exercício 7.

```python
# importando bibliotecas
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# versão bibliotecas
print(f'Versão Pandas: {pd.__version__}')
print(f'Versão Seaborn: {sns.__version__}')
print(f'Versão Matplotlib: {matplotlib.__version__}')
```

```python
# valores do exercício
valores = [155, 142, 149, 130, 151, 163, 151, 142, 156, 133, 138, 161, 128, 144, 172, 137, 151, 166, 147, 163,
           145, 116, 136, 158, 114, 165, 169, 145, 150, 150, 150, 158, 151, 145, 152, 140, 170, 129, 188, 156]
```

```python
# valores ordenados
ordenados = pd.Series(valores).sort_values()
```

```python
# dicionário de classes
classes = {
    'Estrôncio-90 em milibecquerels' : ['110-119', '120-129', '130-139', '140-149', '150-159', '160-169', '170-179', '180-189']
}
```

```python
# categorizar os dados por classes
bins = [110, 120, 130, 140, 150, 160, 170, 180, 190]
cat = pd.cut(x = ordenados, bins = bins, labels = classes['Estrôncio-90 em milibecquerels'], include_lowest = True, right = False)
```

```python
# gerar frequência de classes
freq = cat.value_counts().sort_index()
```

```python
# gerar distribuição de frequência
df = pd.DataFrame(classes)
df['Frequência'] = freq.tolist()
df['Frequência Acumulada'] = df['Frequência'].cumsum()
df
```

```python
# gerar fronteiras de classe
front1 = float(input("Qual a primeira fronteira de classe?"))
ampltd = float(input("Qual a amplitude de classe?"))
front2 = float(input("Qual a última fronteira de classe?"))

ff = []
while front1 <= front2:
    ff.append(round(front1, 2))
    front1+=ampltd
ff
```

```python
# plotando a ogiva

# criando a figura e o eixo
fig, ax = plt.subplots(figsize=(10, 5))

# configurando o eixo
ax.set_facecolor('#E5EAEF')
ax.set_xticks(ff)
ax.set_title("Análise - Quantidade de Estrôncio-90 em milibecquerels")
ax.set_ylabel("Frequência Acumulada")
ax.set_xlabel("Estrôncio-90 em milibecquerels")
ax.set_ylim(0, 45)
ax.grid(visible=True, which='major', axis='both', alpha = 0.7, color = 'white')

# configurando o gráfico de linha
sns.lineplot(x = ff, y = df['Frequência Acumulada'], data = df,
             marker = 'o', markeredgecolor = 'black', markerfacecolor = '#0059B3', markersize = 4,
             linewidth = 0.6, color = '#0059B3', alpha = 0.7)
```
