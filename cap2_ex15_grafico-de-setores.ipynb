### 15. Gráfico de Setores dos Erros dos Candidatos a Emprego
#### Chefes de escritórios financeiros de companhias americanas foram pesquisados sobre as áreas em que os candidatos a emprego cometem erros. Eis as áreas e as frequências das respostas:
- entrevista (452)
- currículo (297)
- carta de apresentação (141)
- verificação de referência (143)
- acompanhamento de entrevista (113)
- telefonema de referência (85).
#### Esses resultados se baseiam em dados de Robert Half Finance and Accounting.
#### Construa um gráfico de setores representando esses dados.

```python
# importando bibliotecas
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Versão bibliotecas
print(f"Versão Pandas: {pd.__version__}")
print(f"Versão Matplotlib: {matplotlib.__version__}")
print(f"Versão Seaborn: {sns.__version__}")
```

```python
# valores do exercício
areas = ['entrevista', 'currículo', 'carta de apresentação', 'verificação de referência', 'acompanhamento de entrevista', 'telefonema de referência']
valores = [452, 297, 141, 143, 113, 85]
```

```python
# deixando os valores ordenados para construção do gráfico
pair = sorted(zip(valores, areas), reverse = True)
valores, areas = zip(*pair)
```

```python
# plotando gráfico de setores
fig, ax = plt.subplots()

#utilizando paleta de cores do seaborn
cores = sns.color_palette('inferno')

ax.pie(x = valores, labels = areas,
      startangle = 90, autopct= '%1.1f%%',
      wedgeprops={'edgecolor': 'black', 'linewidth': 0.2, 'linestyle': '-'},
      textprops={'fontsize': 10}, colors = cores, labeldistance = 1.2, shadow = True)

plt.show()
```
