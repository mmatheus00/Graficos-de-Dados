```python
# função para calcular classes
def classes_dados(qtd_classes, lista):

    # Calcular amplitude de classe
    ordenados = sorted(lista)
    ampltd = round(((max(ordenados) - min(ordenados)) / qtd_classes),2) + 0.01
    print(f'Amplitude de classe: {round(ampltd, 2)}')
    
    # Calcular limites inferiores de classes
    inf = []
    n = ordenados[0]
    for v in range (0, qtd_classes):
        inf.append(round(n, 2))
        n+=ampltd
    
    # Calcular limites superiores de classes
    sup = []
    for v in inf:
        v+=(ampltd-0.01)
        sup.append(round(v, 2))
    
    # Transformar limites inferiores e superiores em string
    inf_str = []
    sup_str = []
    
    for v in inf:
        str(v)
        inf_str.append(str(v))
    
    for v in sup:
        str(v)
        sup_str.append(str(v))

    # Concatenar limites inferiores e superiores para gerar classes
    n = 0
    classe1 = []
    for v in inf_str:
        classe1.append(v + ' - ' + sup_str[n])
        print(f'Classe de dados {n+1}: "{classe1[n]}"', end='\n')
        n+=1
    print(f'\nClasse de dados completa: {classe1}')
```

```python
classes_dados(qtd_classes = 15, lista = valores)
```

```python
import numpy as np
import matplotlib.pyplot as plt

def dotplots(x, x_label="Data Values", x_ticks=None):
    data = np.array(x)
    values, counts = np.unique(data, return_counts=True)

    fig, ax = plt.subplots(figsize=(20, 6))
    for value, count in zip(values, counts):
        ax.plot([value]*count, list(range(count)), 'co', ms=10, linestyle='')

    for spine in ['top', 'right', 'left']:
        ax.spines[spine].set_visible(False)

    ax.yaxis.set_visible(False)
    ax.set_ylim(-1, max(counts) + 1)
    ax.set_xticks(x_ticks if x_ticks else values)
    ax.tick_params(axis='x', length=0, pad=8, labelsize=10)

    plt.suptitle("Diagrama de Pontos")
    plt.ylabel("Frequência")
    plt.xlabel(x_label)

    plt.show()

# Dados: intervalos e frequências
intervalos = [
    (0.15, 0.49), (0.5, 0.84), (0.85, 1.19), (1.2, 1.54), (1.55, 1.89),
    (1.9, 2.24), (2.25, 2.59), (2.6, 2.94), (2.95, 3.29), (3.3, 3.64),
    (3.65, 3.99), (4.0, 4.34), (4.35, 4.69), (4.7, 5.04), (5.05, 5.39)
]
frequencias = [3, 6, 7, 13, 5, 8, 4, 7, 3, 3, 0, 0, 2, 0, 1]

# Gerar lista de dados com repetição conforme frequência
dados = []
for valor, freq in zip(ff, frequencias):
    dados.extend([valor] * freq)

# Gerar gráfico
dotplots(dados, x_label="Descarte de Resíduos Plásticos (lbs)", x_ticks=ff)
```
