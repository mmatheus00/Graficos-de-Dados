### 16. Gráfico de Setores dos Erros dos Candidatos a Emprego
#### Construa um gráfico de Pareto dos dados apresentados no Exercício 15.
#### Compare o gráfico de Pareto com o gráfico de setores.

```python
# importando bibliotecas
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Versão bibliotecas
print(f"Versão Pandas: {pd.__version__}")
print(f"Versão Matplotlib: {matplotlib.__version__}")
print(f"Versão Seaborn: {sns.__version__}")
```

```python
# valores do exercício
areas = ['entrevista', 'currículo', 'carta de apresentação', 'verificação de referência', 'acompanhamento de entrevista', 'telefonema de referência']
valores = [452, 297, 141, 143, 113, 85]
```

```python
# deixando os valores ordenados para construção do gráfico
pair = sorted(zip(valores, areas), reverse = True)
valores, areas = zip(*pair)
```

```python
# plotando gráfico de pareto
sns.catplot(x = areas, y = valores, kind="bar",
           edgecolor = "black", width = 0.5, palette='inferno', hue = areas)

sns.color_palette(palette='inferno')
plt.xticks(areas, ha = 'right', rotation = 45)
plt.ylim(0, 500)
plt.show()
```

### Qual gráfico é mais eficaz para mostrar a importância relativa dos erros cometidos por pessoas que procuram emprego?

- O gráfico de Pareto. Porque ele identifica e prioriza visualmente os principais problemas ou causas que geram a maioria dos resultados ou consequências.- Além disso, ele também garante melhor visualização quando temos muitas variáveis categóricas.
