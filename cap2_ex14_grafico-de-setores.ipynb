### 14. Gráfico de Setores para Matrículas na Graduação
#### Construa um gráfico de setores para os dados na Tabela 2-9. Compare o gráfico de setores com o gráfico de Pareto no Exercício 13.

```python
# importando bibliotecas
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Versão bibliotecas
print(f"Versão Pandas: {pd.__version__}")
print(f"Versão Matplotlib: {matplotlib.__version__}")
print(f"Versão Seaborn: {sns.__version__}")
```

```python
# valores tabela 2-9
faculdade = ["Pública de 2 anos", "Pública de 4 anos", "Particular de 2 anos", "Particular de 4 anos"]
freq = [36.8, 40.0, 1.6, 21.9]
```

```python
# deixando os valores ordenados para construção do gráfico
pair = sorted(zip(freq, faculdade), reverse = True)
freq, faculdade = zip(*pair)
```

```python
# plotando gráfico de setores
fig, ax = plt.subplots()

#utilizando paleta de cores do seaborn
cores = sns.color_palette('inferno')

ax.pie(x = freq, labels = faculdade,
      startangle = 90, autopct= '%1.1f%%',
      wedgeprops={'edgecolor': 'black', 'linewidth': 0.2, 'linestyle': '-'},
      textprops={'fontsize': 10}, colors = cores,
      labeldistance = 1.2, shadow = True)

plt.show()
```
